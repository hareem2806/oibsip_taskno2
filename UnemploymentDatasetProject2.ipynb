# **1. Visualizing Unemployment Rate**

*   By Monthly Average
*   Subsetting by Months
*   Identifying State with Highest Unemployment Rate






# By Monthly Average

```python
monthly_avg = df.groupby('Month')['Estimated Unemployment Rate(%)'].mean().reset_index()
sns.barplot(x='Month', y='Estimated Unemployment Rate(%)', data=monthly_avg)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Average of Unemployment Rate by Months ')
plt.show()
```

***Through this bar plot we observe that the months of April and May suffer highest unemployment rate on average***

# Subsetting by Months


 ***Subsetting for the Months which have the highest estimation of Unemployment Rate as observed from the previous graph ***

```python
# Filter the DataFrame for the desired months
selected_months = ['Apr', 'May']
filtered_df = df[df['Month'].isin(selected_months)]

plt.figure(figsize=(15, 5))

sns.barplot(x='States', y='Estimated Unemployment Rate(%)', data=filtered_df, hue='Month', width=0.8)
plt.xlabel('States', fontsize=15)
plt.ylabel('Unemployment Rate', fontsize=15)
plt.title('States vs Unemployment Rate According to Months' , fontsize=15)

plt.xticks(rotation=90)

plt.show()
```

***Through this visualization of unenemployment rates for the months of April and May in different states we observe that Puducherry suffers highest unemployment rate***

# Identifying State with Highest Unemployment Rate

***As we have seen that Puducherry has the highest Unemployment Rate so we will visualize it in detail ***

```python
# Subset the data for the state of 'Puducherry'

subset_df = df[df['States'] == 'Puducherry']

sns.barplot(x='Month', y='Estimated Unemployment Rate(%)', data=subset_df)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate in Puducherry by Month')
plt.show()
```

# **2. Visualizing Estimation Of Employeed People**

*   By Monthly Average
*   Subsetting by Months
*   Identifying State with Highest Employment






# By Monthly Average

```python
monthly_avg = df.groupby('Month')['Estimated Employed'].mean().reset_index()
sns.barplot(x='Month', y='Estimated Employed', data=monthly_avg)
plt.xlabel('Month')
plt.ylabel('Estimated Employed')
plt.title('Average of Estimated Employees by Months ')
plt.show()
```

***With this bar plot we observe that on average the months of Jan, Feb, Mar, Jul and Aug have the highest estimation of people employed***

# Subsetting by Months

 ***Subsetting for the Months which have the highest estimation of people employed as observed from the previous graph ***

```python
# Filter the DataFrame for the desired months
selected_months = ['Jan', 'Feb', 'Mar', 'Jul', 'Aug']
filtered_df = df[df['Month'].isin(selected_months)]

plt.figure(figsize=(30, 10))
sns.barplot(x='States', y='Estimated Employed', data=filtered_df,  hue='Month', width=1.0)
plt.xlabel('States', fontsize=20)
plt.ylabel('Unemployment Employed', fontsize=20)
plt.title('States vs. Estimated Employed According to Months' , fontsize=20)

plt.xticks(rotation=90, fontsize=20)

plt.show()
```

***Through this visualization we observe that Uttar Pradesh has the highest number of people employed for the months of Jan, Feb, Mar, Jul and Aug***

# Identifying State with Highest Employment

***As we have seen that Uttar Pradesh has the highest number of people employed so we will visualize it in detail ***

```python
# Subset the data for the state of 'Uttar Pradesh'
subset_df = df[df['States'] == 'Uttar Pradesh']

sns.barplot(x='Month', y='Estimated Employed People', data=subset_df)
plt.xlabel('Month')
plt.ylabel('Estimated Employed People')
plt.title('Estimated Employeed People in Uttar Pradesh  by Month')

plt.show()
```

# **3. Visualizing Estimated Labour Participation Rate**

*   By Monthly Average
*   Subsetting by Months
*   Identifying State with Highest Labour Participation Rate






#By Monthly Average

```python
monthly_avg = df.groupby('Month')['Estimated Labour Participation Rate(%)'].mean().reset_index()
sns.barplot(x='Month', y='Estimated Labour Participation Rate(%)', data=monthly_avg)
plt.xlabel('Month')
plt.ylabel('Estimation of Labour Participation Rate(%)')
plt.title('Average estimation of Labour Participation rate by Months')
plt.show()
```

***With this bar plot we observe that on average the months of Jan, Feb, Mar and Aug have the highest estimation of Labour Participation Rate***

# Subsetting by Months

 ***Subsetting for the Months which have the highest estimation of Labour Participation Rate as observed from the previous graph ***

```python
# Filter the DataFrame for the desired months
selected_months = ['Jan', 'Feb', 'Mar','Aug', 'Jul']
filtered_df = df[df['Month'].isin(selected_months)]

plt.figure(figsize=(15, 8))
sns.barplot(x='States', y='Estimated Labour Participation Rate(%)', data=filtered_df, hue='Month', width=0.9)
plt.xlabel('States', fontsize=16)
plt.ylabel('Estimated Labour Participation Rate(%)', fontsize=16)
plt.title('States vs Estimated Labour Participation Rate(%) for Jan, Feb, March and Aug', fontsize=15)
plt.xticks(rotation=90, fontsize=12)
plt.show()
```

***Through this barplot we observe that the stae Tripura has the highest labour participation rate followed by Meghalaya and Telangana states***

# Identifying State with Highest Labour Participation Rate

***As we have seen that Tripura has the highest Labour Participation Rate so we will visualize it in detail ***

```python

# Subset the data for the state of 'Tripura'
subset_df = df[df['States'] == 'Tripura']

sns.barplot(x='Month', y='Estimated Labour Participation Rate(%)', data=subset_df)
plt.xlabel('Months')
plt.ylabel('Estimated Labour Participation Rate(%)')
plt.title('Estimated Labour Participation Rate(%) in Tripura by Month')

plt.show()
```
